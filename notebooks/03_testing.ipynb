```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# --- STEP 1: LOAD AND PREPARE ---
# Load the dataset you prepared earlier
df = pd.read_csv('prepared_analysis.csv')

# Recreate the User-Movie Matrix
# Rows = Users, Columns = Movie Titles, Values = Ratings
user_movie_matrix = df.pivot_table(index='userId', columns='title', values='rating').fillna(0)

print(f"✅ Step 1 Complete: Matrix created with shape {user_movie_matrix.shape}")

# --- STEP 2: COMPUTE SIMILARITY ---
# We transpose (.T) because we want to compare MOVIES (columns), not users
movie_similarity_matrix = cosine_similarity(user_movie_matrix.T)

print("✅ Step 2 Complete: Cosine Similarity computed.")

# --- STEP 3: CONVERT TO SEARCHABLE DATAFRAME ---
movie_similarity_df = pd.DataFrame(
    movie_similarity_matrix, 
    index=user_movie_matrix.columns, 
    columns=user_movie_matrix.columns
)

print("✅ Step 3 Complete: Similarity DataFrame is ready.")

# --- STEP 4: DEFINE SMART RECOMMENDATION FUNCTION ---
def get_recommendations(target_title):
    # 1. Handle potential title mismatches (Case-insensitive & Partial match)
    all_titles = movie_similarity_df.columns
    match = [t for t in all_titles if target_title.lower() in t.lower()]
    
    if not match:
        return "❌ Movie not found. Please check the spelling."
    
    # Use the first best match found
    actual_title = match[0]
    print(f"\nTarget found: '{actual_title}'")
    
    # 2. Get similarity scores
    scores = movie_similarity_df[actual_title].sort_values(ascending=False)
    
    # 3. Filter out the movie itself and take top 10
    recommendations = scores.iloc[1:11]
    
    return recommendations

# --- STEP 5: VERIFY WITH TESTS ---
print("\n--- TEST 1: FAMILY/ANIMATION ---")
print(get_recommendations("Toy Story"))

print("\n--- TEST 2: SCI-FI/CLASSIC ---")
print(get_recommendations("2001: A Space Odyssey"))
```
